```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\hp\Downloads\airquality.csv", sep=',')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

# data cleaning

```python
df = df.drop('Unnamed: 0', axis = 1)
df
```

# filling null values

```python
df['Ozone'] = df['Ozone'].fillna(df['Ozone'].mean())
df['Solar.R'] = df['Solar.R'].fillna(df['Solar.R'].mean())
df
```

```python
df.isnull().sum()
```

# data transformation for humidity column

# data visualization

# 1 Scatter Plot

```python
plt.scatter(x = df['Ozone'], y = df['Temp'])
plt.legend(["Ozone"])
plt.title("Ozone Scatter Plot")
plt.xlabel("Ozone")
plt.ylabel("Temp")
plt.show()                                   #matplotlib library
```

# 2.Bar Plot

```python
sns.barplot(x = df['Month'] , y = df['Temp'])                 #seaborn library
```

# 3.Heatmap

```python
co = df.corr()
sns.heatmap(co, vmin = -1, vmax = 1, annot = True)
```

# 4.PairPlot

```python
sns.pairplot(df)
```

# 5.LineGraph

```python
h = df.iloc[1:15, 0]                #0 because of ozone column
v = df.iloc[1:15, 3]                #3 because of temp column
plt.plot(h, label = "Ozone", marker = "o", linestyle = "dotted")
plt.plot(v, label = "Temp", marker = "o", linestyle = "dashed")
plt.title("Line Graph for Ozone and Temp")
plt.legend()
plt.show()
```

# 6.BoxPlot

```python
sns.boxplot(x = df['Month'], y = df['Temp'])
```

# 7.PieChart

```python
lables = ['Ozone','Solar.R','Wind','Temp']
sizes = [df['Ozone'].mean(), df['Solar.R'].mean(), df['Wind'].mean(), df['Temp'].mean()]
colors = ['red', 'blue','yellow', 'green']
textprops = {"fontsize" : 15}
plt.pie(sizes, labels = lables, colors = colors, autopct = '%1.1f%%', shadow = True)
plt.title("Factors", fontsize = 20, style = "italic")
```

# 8.Histogram

```python
h = df.iloc[:, -1]
plt.hist(h, bins = 'auto')
plt.title('Histogram')
plt.xlabel('Ozone')
plt.ylabel('Temperature')
```
